```python
#乘客总共有11个特征属性，预测：“什么样的人更有可能生存？”。
# PassengerId => 乘客ID
# Pclass => 客舱等级(1/2/3等舱位)
# Name => 乘客姓名
# Sex => 性别
# Age => 年龄
# SibSp => 兄弟姐妹数/配偶数
# Parch => 父母数/子女数
# Ticket => 船票编号
# Fare => 船票价格
# Cabin => 客舱号
# Embarked => 登船港口

import pandas as pd
```

```python
train=pd.read_csv('./train.csv')
```

```python
train
```

```python
test=pd.read_csv('./test.csv')
```

```python
test
```

```python
full=pd.concat([train,test],ignore_index=True,sort = True)
```

```python
full.head()
```

```python
full.isnull().sum()
```

```python
full.Embarked.mode()  #求出众数
```

```python
full['Embarked'].fillna('S',inplace=True)  #登船港口 众数S填充
```

```python
full[full.Fare.isnull()]
```

```python
full.Fare.fillna(full[full.Pclass==3]['Fare'].median(),inplace=True) #船票价格 Pclass3中位数填充
```

```python
full.loc[full.Cabin.notnull(),'Cabin']=1
full.loc[full.Cabin.isnull(),'Cabin']=0

#性别清洗
full.loc[full['Sex']=='male','Sex']=1
full.loc[full['Sex']=='female','Sex']=0
```

```python
full['Title']=full['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
full.Title.value_counts()
```

```python
nn={'Capt':'Rareman', 'Col':'Rareman','Don':'Rareman','Dona':'Rarewoman',
    'Dr':'Rareman','Jonkheer':'Rareman','Lady':'Rarewoman','Major':'Rareman',
    'Master':'Master','Miss':'Miss','Mlle':'Rarewoman','Mme':'Rarewoman',
   'Mr':'Mr','Mrs':'Mrs','Ms':'Rarewoman','Rev':'Mr','Sir':'Rareman',
   'the Countess':'Rarewoman'}
full.Title=full.Title.map(nn)

full.loc[full.PassengerId==797,'Title']='Rarewoman' #女博士划分为Rarewoman
```

```python
full.Title.value_counts()
```

```python
full[full.Title=='Master']['Sex'].value_counts()
```

```python
full[full.Title=='Master']['Age'].describe()
```

```python
full[full.Title=='Miss']['Age'].describe()
```

```python
full.Age.fillna(999,inplace=True) #年龄缺失值填充
```

```python
def girl(aa):
    if (aa.Age!=999)&(aa.Title=='Miss')&(aa.Age<=14):
        return 'Girl'
    elif (aa.Age==999)&(aa.Title=='Miss')&(aa.Parch!=0):
        return 'Girl'
    else:
        return aa.Title

full['Title']=full.apply(girl,axis=1)
```

```python
full.Title.value_counts()
```

```python
full[full.Age==999]['Age'].value_counts()
```

```python
Tit=['Mr','Miss','Mrs','Master','Girl','Rareman','Rarewoman']
for i in Tit:
    full.loc[(full.Age==999)&(full.Title==i),'Age']=full.loc[full.Title==i,'Age'].median()
```

```python
#年龄处理
full.loc[full['Age']<=15,'Age']=0
full.loc[(full['Age']>15)&(full['Age']<55),'Age']=1
full.loc[full['Age']>=55,'Age']=2
```

```python
#客舱等级(0/1/2)
full['Pclass']=full['Pclass']-1
```

```python
full.info()
```

```python
from sklearn.cluster import KMeans
```

```python
Fare=full['Fare'].values
Fare=Fare.reshape(-1,1)
km = KMeans(n_clusters=3).fit(Fare)  #聚类为3类
Fare = km.fit_predict(Fare)
full['Fare']=Fare
```

```python
full['Fare']=full['Fare'].astype(int)
full['Age']=full['Age'].astype(int)
full['Cabin']=full['Cabin'].astype(int)
full['Pclass']=full['Pclass'].astype(int)
full['Sex']=full['Sex'].astype(int)
```

```python
dataset=full.drop(columns=['Embarked','Name','Parch','PassengerId','SibSp','Ticket','Title'])
dataset.dropna(inplace=True)
dataset['Survived']=dataset['Survived'].astype(int)
train_res=dataset[:800]
test_res=dataset[800:892]
```

```python
from  pgmpy.estimators import ParameterEstimator
from pgmpy.models import BayesianModel
from pgmpy.estimators import BayesianEstimator
from pgmpy.inference import VariableElimination
# from matplotlib import pyplot as plt
# from pgmpy.estimators import MaximumLikelihoodEstimator
import warnings
warnings.filterwarnings("ignore")
```

```python
model = BayesianModel([('Age', 'Survived'), ('Sex', 'Survived'),('Fare','Pclass'),('Pclass','Survived'),('Cabin','Survived')])
```

```python
model.edges()
```

```python
def showBN(model,save=False):
    from graphviz import Digraph
    node_attr = dict(
     style='filled',
     shape='box',
     align='left',
     fontsize='12',
     ranksep='0.1',
     height='0.2'
    )
    dot = Digraph(node_attr=node_attr, graph_attr=dict(size="12,12"))
    seen = set()
    edges=model.edges()
    for a,b in edges:
        dot.edge(a,b)
    if save:
        dot.view(cleanup=True)
    return dot 
```

```python
showBN(model)
```

```python
model.fit(train_res, estimator=BayesianEstimator, prior_type="BDeu")
```

```python
# 验证模型数据的正确性（检测节点是否定义，概率和是否为1）
model.check_model()
```

```python
for i in model.get_cpds():
    print(i)
```

```python
model_infer = VariableElimination(model)
```

```python
#边缘概率
for i in model_infer.query(['Age', 'Survived', 'Sex','Fare','Pclass','Cabin']).values():
    print(i)
```

```python
#推理
```

```python
q1 = model_infer.query(variables=['Survived'], evidence={'Fare': 0})
print(q1['Survived'])
```

```python
q2 = model_infer.map_query(variables=['Fare','Age','Sex','Pclass','Cabin'], evidence={'Survived': 1})
# 此处的因果关系并不可以解释为一般意义上的逻辑因果，而是表示一种概率上的相关
print(q2)  #{'Sex': 0, 'Fare': 0, 'Age': 1, 'Pclass': 2, 'Cabin': 0}
```

```python
#剔除想预测的列输入predict函数
predict_data=test_res.drop(columns=['Survived'],axis=1)
y_pred = model.predict(predict_data)
print("测试精度：{}".format((y_pred['Survived']==test_res['Survived']).sum()/len(test_res)))
```

```python
#基于评分的结构学习
from pgmpy.estimators import HillClimbSearch
from pgmpy.estimators import BicScore,BdeuScore,K2Score
hc_bd_model = HillClimbSearch(train_res, scoring_method=BdeuScore(train_res)).estimate()
hc_bi_model = HillClimbSearch(train_res, scoring_method=BicScore(train_res)).estimate()
hc_k2_model = HillClimbSearch(train_res, scoring_method=K2Score(train_res)).estimate()
```

```python
def test_predict(test_model):
    test_model.fit(train_res, estimator=BayesianEstimator, prior_type="BDeu") #  equivalent_sample_size=5
    predict_data=test_res.drop(columns=['Survived'],axis=1)
    y_pred = test_model.predict(predict_data)
    print((y_pred['Survived']==test_res['Survived']).sum()/len(test_res))
    
test_predict(hc_bd_model)
```
